```
def tambah_matriks(matriks1, matriks2):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks1[i][j] + matriks2[i][j]
    return hasil

def kurang_matriks(matriks1, matriks2):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks1[i][j] - matriks2[i][j]
    return hasil

def kali_matriks(matriks1, matriks2):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            for k in range(2):
                hasil[i][j] += matriks1[i][k] * matriks2[k][j]
    return hasil

def bagi_matriks(matriks1, matriks2):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            if matriks2[i][j] != 0:
                hasil[i][j] = matriks1[i][j] / matriks2[i][j]
            else:
                print("Tidak bisa melakukan pembagian, karena ada pembagian dengan nol.")
                return None
    return hasil

def tambah_konstanta(matriks, konstanta):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks[i][j] + konstanta
    return hasil

def kurang_konstanta(matriks, konstanta):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks[i][j] - konstanta
    return hasil

def kali_konstanta(matriks, konstanta):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks[i][j] * konstanta
    return hasil

def bagi_konstanta(matriks, konstanta):
    hasil = [[0, 0], [0, 0]]
    for i in range(2):
        for j in range(2):
            hasil[i][j] = matriks[i][j] / konstanta
    return hasil

def input_matriks():
    matriks = []
    for i in range(2):
        row = []
        for j in range(2):
            elemen = float(input(f"Masukkan elemen matriks baris ke-{i+1} kolom ke-{j+1}: "))
            row.append(elemen)
        matriks.append(row)
    return matriks

def cetak_matriks(matriks):
    for row in matriks:
        print(row)

# Input matriks pertama
print("Masukkan matriks pertama:")
matriks1 = input_matriks()

# Input matriks kedua
print("Masukkan matriks kedua:")
matriks2 = input_matriks()

# Tambahkan matriks
hasil_penambahan = tambah_matriks(matriks1, matriks2)
print("Hasil penambahan matriks:")
cetak_matriks(hasil_penambahan)
konstanta_tambahan1 = 50

# Tambahkan konstanta ke matriks pertama
hasil_tambah_konstanta = tambah_konstanta(matriks1, konstanta_tambahan1)
print(f"Hasil penambahan matriks pertama dengan konstanta {konstanta_tambahan}:")
cetak_matriks(hasil_tambah_konstanta)

# Kurangkan matriks (A - B)
hasil_pengurangan1 = kurang_matriks(matriks1, matriks2)
print("Hasil pengurangan matriks (A - B):")
cetak_matriks(hasil_pengurangan1)

# Kurangkan matriks (B - A)
hasil_pengurangan2 = kurang_matriks(matriks2, matriks1)
print("Hasil pengurangan matriks (B - A):")
cetak_matriks(hasil_pengurangan2)

# Kurangkan konstanta dari matriks kedua
hasil_kurang_konstanta = kurang_konstanta(matriks1, konstanta_tambahan1)
print(f"Hasil pengurangan matriks kedua dengan konstanta {konstanta_tambahan}:")
cetak_matriks(hasil_kurang_konstanta)

# Kalikan matriks
hasil_perkalian = kali_matriks(matriks1, matriks2)
print("Hasil perkalian matriks:")
cetak_matriks(hasil_perkalian)

konstanta_tambahan2 = 2.5

# Kalikan matriks pertama dengan konstanta
hasil_kali_konstanta = kali_konstanta(matriks1, konstanta_tambahan2)
print(f"Hasil perkalian matriks pertama dengan konstanta {konstanta_tambahan}:")
cetak_matriks(hasil_kali_konstanta)

# Bagi matriks
hasil_pembagian = bagi_matriks(matriks1, matriks2)
if hasil_pembagian:
    print("Hasil pembagian matriks:")
    cetak_matriks(hasil_pembagian)

# Bagi matriks pertama dengan konstanta
hasil_bagi_konstanta = bagi_konstanta(matriks1, konstanta_tambahan2)
print(f"Hasil pembagian matriks pertama dengan konstanta {konstanta_tambahan}:")
cetak_matriks(hasil_bagi_konstanta)
```
